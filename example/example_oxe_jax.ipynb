# OXE JAX Dataset Example

This notebook shows how to use `OXEJAXDataset` for the OXE-only JAX path.
Numeric modalities are returned as JAX arrays, while text fields stay as Python objects.


## Prerequisites

Install the OXE extras plus JAX in your environment before running this notebook.

```bash
pip install -e '.[oxe]'
pip install jax
```


````python
import numpy as np
import jax
from robotdataset import OXEJAXDataset, list_datasets
````

## 1. Discover datasets


````python
datasets = list_datasets()
list(datasets.items())[:5]
````

## 2. Load an OXE dataset through the JAX path

This example streams two episodes and requests temporal windows for image/state/action fields.


````python
dataset = OXEJAXDataset(
    dataset_name="cmu_playing_with_food",
    split="train",
    episodes=[0, 2],
    batch_size=4,
    delta_timestamps={
        "observation/image": [-0.2, -0.1, 0.0],
        "observation/state": [-0.2, -0.1, 0.0],
        "action": [0.0, 0.1],
    },
    control_frequency=10.0,
)
dataset
````

````python
dataset.modalities
````

````python
dataset.image_keys
````

## 3. Sample a JAX batch


````python
batch = dataset.get_batch()
type(batch), batch.keys()
````

````python
print(type(batch["observation"]["image"]))
print(batch["observation"]["image"].shape)
print(batch["next"]["observation"]["image"].shape)
print(batch["observation"]["state"].shape)
print(batch["action"].shape)
````

````python
print(batch["observation"]["image"].dtype)
print(batch["action"].dtype)
print(type(batch["language_instruction"]))
print(batch["language_instruction"][:2])
````

## 4. Convert back to NumPy when needed

JAX arrays can be materialized as NumPy arrays for downstream tools that are not JAX-native.


````python
image_np = np.asarray(batch["observation"]["image"])
state_np = np.asarray(batch["observation"]["state"])
print(type(image_np), image_np.shape)
print(type(state_np), state_np.shape)
````

````python
print(jax.tree.map(lambda x: getattr(x, "shape", None), batch))
````
